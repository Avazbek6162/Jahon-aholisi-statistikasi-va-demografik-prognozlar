```python
!pip install wbdata
import wbdata
import pandas as pd
import matplotlib.pyplot as plt
```

## **2-qadam. Indikatorni tanlash**

Jahon aholisi soni statistikasi uchun Jahon banki indikatoridan foydalanamiz: `SP.POP.TOTL`.

```python
# Jahon aholisi soni uchun indikator
indicator = {"SP.POP.TOTL": "Aholi soni"}
```

## **3-qadam. Maâ€™lumotlarni yuklab olish**

`wbdata.get_dataframe()` yordamida jahon aholisi maâ€™lumotlarini yuklab olamiz.

```python
# Jahon boâ€˜yicha maâ€™lumotlarni yuklash
df = wbdata.get_dataframe(indicator, country="WLD")
```

## **4-qadam. Vaqt indeksiga oâ€˜tkazish**

Maâ€™lumotlarni `datetime` formatiga oâ€˜tkazamiz va yillar boâ€˜yicha tartiblaymiz.

```python
# Indeksni sana formatiga oâ€˜tkazish
df.index = pd.to_datetime(df.index)

# Yillar boâ€˜yicha tartiblash
df = df.sort_index()
```

## **5-qadam. Yillar oraliqni tanlash**

1960-yildan 2025-yilgacha boâ€˜lgan davrni ajratib olamiz.

```python
# Faqat 1960â€“2025 yillarni ajratib olish
df = df.loc["1960":"2025"]
```

## **6-qadam. Grafik chizish**

Jahon aholisi sonini milliardlarda tasvirlaymiz.

```python
# Grafik chizish
plt.figure(figsize=(10, 6))
plt.plot(df.index.year, df["Aholi soni"] / 1e9, marker="o", color="blue")
plt.title("Jahon aholisi statistikasi (1960â€“2025)", fontsize=14)
plt.xlabel("Yil")
plt.ylabel("Aholi soni (mlrd)")
plt.grid(True)
plt.show()
```
